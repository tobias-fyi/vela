## Basic Transformations

### Cropping

As an example to demonstrate how to crop an image using Pillow, I'm going to create what will be the header image for a blog post called _Automated Image Background Removal with Python_.

The original image is tall (600 x 900). I need to crop it to be wide (600 x 314) to make it show up correctly when posted to social media (specifically LinkedIn).

```python
# === Download the original image === #
# NOTE: remove the "assets/" if not using an assets subdirectory
!wget https://raw.githubusercontent.com/tobias-fyi/assetstash/master/visual/images/img_seg_bin_mask/06_paws.jpg -q -O assets/paws.jpg
```

```python
# === Load and display original image === #
paws = Image.open("assets/paws.jpg")
print("Image dimensions:", paws.size)
plt.imshow(paws);
```

```python
# === Crop === #

# Define the rectangle to crop
top = 280
hgt = 314
box = (0, top, 600, top + hgt)

# Make the crop
paws_crop = paws.crop(box)

# Look at the result
print(paws_crop.size)
plt.imshow(paws_crop);
```

```python
# === Save the cropped image to local filesystem === #
paws_crop.save("assets/paws_header_crop.jpg")
```
